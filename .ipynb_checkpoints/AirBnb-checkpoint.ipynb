●	Review for possible duplicates and erroneous data.

```python
# reformate price function
def dollar_to_float(value):
    try:
        return float(value.strip('$').strip(','))
    
    except:
        return np.nan
    
```

```python
# using dollar_to_float function to remformate price col. 
AirbnbDS['price'] = AirbnbDS['price'].apply(dollar_to_float)
```

```python
# check the result
AirbnbDS['price'].head()
```

```python
# number of  rows before duplicates.
drows = AirbnbDS['id'].value_counts().sum()
print ('number of rows before duplicates: ',drows)
```

```python
# remove the duplicated of 'id'
AirbnbDS['id'] = AirbnbDS['id'].drop_duplicates(keep='first',inplace=False)
```

### PROJECT STEP 2: DATA EXPLORATION AND ANALYSIS (BY PROMPT)

**PROMPT 1: How much revenue do successful hosts make?**

To answer this question we have to do the following: 

``1. Calculate a proxy number of stays for each listing by assuming that 50 percent of customers who stayed left a review ``

``2. Calculate estimated daily revenue for two guests!`` 

``3. Multiply the estimated daily revenue by the minimum number of nights !``

``4. Calculate an estimated_total_revenue for each listing!``

columns that we dealing with our calculation:
>  [ price, guests_included, extra_people, accommodates ]


```python
# function to get number of Stays for eaching listing by assuming 50% of gusets stary who left review 
def NoOfStays(value):
    return int(value * 2)
```

```python
AirbnbDS['Stays'] = AirbnbDS['number_of_reviews'].apply(NoOfStays)
```

```python
AirbnbDS['Stays'].head()
```

Calculate estimated daily revenue for two guests

```python
# check the formate of extra_people
AirbnbDS['extra_people'].head()
```

we found above the formate of extra_people neeed to reformating using the following: 

```python
# using pandas filtering method to extract $ from vlaue of extra_people to do our calculation to get 
# estimated_daily_revenue
AirbnbDS['extra_people'] = AirbnbDS['extra_people'].str.extract('(\d+)')
```

```python
AirbnbDS['extra_people'] = AirbnbDS['extra_people'].map(lambda x : float(x))
```

```python
# check the formate of extra_people after applying the reformatting 
AirbnbDS['extra_people'].head()
```

```python
# 2. calculate the estimated daily revenue
# add new columns for daily revenue 
AirbnbDS['revenue'] = AirbnbDS['price']
```

```python
# function to calculate the estimated daily revenue
def estimated_daily_revenue(row):
    if row['guests_included'] > 1:
        return row['revenue']
    elif row['guests_included'] == 1 and row['accommodates'] == 1:
        return row['revenue']
    elif row['guests_included'] == 1 and row['accommodates'] > 1:
        return row['revenue'] + row['extra_people']
    else :
        row['revenue'] = None # this condition if there is 0 
        return row['revenue']
  
```

```python
# apply our calculation 
EDR = AirbnbDS.apply(estimated_daily_revenue,axis=1)
```

```python
# figure out the data 
AirbnbDS[['revenue','price','guests_included','accommodates','extra_people']]
```

```python
# figure out the estimated daily revenue
EDR
```

```python
# refill NAN fields by the median of estimated_daily_revenue
EDR = AirbnbDS['revenue'] = AirbnbDS['revenue'].map(lambda x: np.median if x == 'NaN' else float(x))
```

```python
AirbnbDS[['revenue','price','guests_included','accommodates','extra_people']].head()
```

```python
# estimated daily revenue
EDR.head()
```

```python
#3. Multiply the estimated daily revenue by the minimum number of nights !
#Estimated_rev_per_booking
AirbnbDS['revenue_per_night'] = AirbnbDS['revenue'] * AirbnbDS['minimum_nights']
```

```python
# 4. Calculate an estimated_total_revenue for each listing!
# calculate the total ER (estimated revenue )
AirbnbDS['estimated_total_revenue']= AirbnbDS['revenue_per_night'] * AirbnbDS['Stays']
```

```python
# figure out the resutls as follow. 
AirbnbDS[['revenue','minimum_nights', 'revenue_per_night','Stays','estimated_total_revenue']].head(10)
```
